# **PROBLEM 2**

# **The breaking strength of gunny bags used for packaging cement is normally distributed with a mean of 5 kg per sq. centimeter and a standard deviation of 1.5 kg per sq. centimeter. The quality team of the cement company wants to know the following about the packaging material to better understand wastage or pilferage within the supply chain.**

Here we will use the normal distribution

Formulas:

cdf (cumulative density function) = cdf(x, mu(μ), sigma(σ))

pdf (probability density function) = pdf(x, mu(μ), sigma(σ))


Where,

x = observed value

mu(μ) = mean

sigma(σ) = standard deviation

### **Importing all the required libraries**

```python
# Libraries to help with reading and manipulating data
import numpy as np
import pandas as pd

# Libraries to help with data visualization
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Library to help with statistical analysis
import scipy.stats as stats
from scipy.stats import norm
```

### **<a name='link7'>2.1 What proportion of the gunny bags have a breaking strength of less than 3.17 kg per sq cm?**

```python
mu = 5
sigma = 1.5

# the cumulative probability

probability_less_than_3_17 = norm.cdf(3.17, mu, sigma)

print("The probability that proportion of the gunny bags have a breaking strength of less than 3.17 kg per sq cm is:",round(probability_less_than_3_17, 4))
```

```python
# plot the pdf of breaking strength using norm.pdf()

density = pd.DataFrame()  # create an empty DataFrame

density["x"] = np.linspace(1, 10, 500)  # create an array of 500 numbers in between the 1 and 10.

density["pdf"] = norm.pdf(density["x"], 5, 1.5)  # calculate the pdf() of the create density["x"]

plt.plot(density["x"], density["pdf"]) # plot the pdf of the normal distribution

plt.axvline(x=3.17, c="g") # draw a red vertical line at x = 3.17

x1 = np.linspace(density["x"].min(), 3.17, 10) # create an array of 10 numbers between density["x"].in() and 3.17
plt.fill_between(x1, norm.pdf(x1, 5, 1.5), color="y") # fill the area under the specified region with yellow color
plt.xlabel("Breaking Strength") # set the x-axis label
plt.ylabel("Probability") # set the y-axis label
plt.title("Fig 1: Breaking strength of less than 3.17 kg per sq cm") # set the title
plt.show()
```

### **<a name='link7'>2.2 What proportion of the gunny bags have a breaking strength of at least 3.6 kg per sq cm.?**

```python
mu = 5
sigma = 1.5

# the cumulative probability

probability_more_than_3_6 = 1 - norm.cdf(3.6, mu, sigma)

print("The probability that proportion of the gunny bags have a breaking strength of at least 3.6 kg per sq cm is:",round(probability_more_than_3_6, 4))
```

```python
# plot the pdf of breaking strength using norm.pdf()

density = pd.DataFrame() # create an empty DataFrame
density["x"] = np.linspace(1, 10, 500) # create an array of 500 numbers in between the 1 and 10.
density["pdf"] = norm.pdf(density["x"], 5, 1.5) # calculate the pdf() of the create density["x"]
plt.plot(density["x"], density["pdf"]) # plot the pdf of the normal distribution
plt.axvline(x=3.6, c="g") # draw a red vertical line at x = 3.6
x1 = np.linspace(3.6, density["x"].max(), 200) # create an array of 10 numbers between 3.6 and density["x"].max()
plt.fill_between(x1, norm.pdf(x1, 5, 1.5), color="y") # fill the area under the specified region with yellow color
plt.xlabel("Breaking Strength") # set the x-axis label
plt.ylabel("Probability") # set the y-axis label
plt.title("Fig 2: Breaking strength of at least 3.6 kg per sq cm") # set the title
plt.show()
```

### **<a name='link7'>2.3 What proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.?**

## **<a name='link7'>3.1 Zingaro has reason to believe that the unpolished stones may not be suitable for printing. Do you think Zingaro is justified in thinking so?**



The hypothesis are as below:

**Null Hypothesis:**Unpolished stone is suitable for printing which means Brinell's hardness index is atleast or equal to 150.

**Alternative Hypothesis:**Unpolished stone is not suitable for printing which means Brinell's hardness index is less than 150.



*   Ho: µ >= 150
*   Ha: µ  < 150

      Where µ be the population mean of unpolished stones.

Level of Significance: α = 0.05

Sample Size = 75





Now, we will calculate the value of T-statistics and p-value, and compare them to reject OR fail to reject the Null hypothesis.

We will use the independent one sample one tailed t-test to calculate T_stats and p-value.

```python
test_stat, p_value = stats.ttest_1samp(df['Unpolished '], popmean = 150, alternative = 'less')
print('The p-value is ', p_value)
print('The test_stat is', test_stat)
```

**Conclusion:**

The p-value(4.171286997419652e-05) is less than the level of significance. so we can reject the null hypothesis with 95% confidence, we can say that we have enough evidence to say that the hardness for unpolished stones is less than 150.

Thus Zingaro is right in thinking that polished stones are not fit for painting.

## **<a name='link7'>3.2 Is the mean hardness of the polished and unpolished stones the same?**

The hypothesis are as below:

**Null Hypothesis:**The mean hardness of the polished and unpolished stones the same.

**Alternative Hypothesis:**The mean hardness of the polished and unpolished stones are not the same.

Ho: µ(Unpolished) = µ(Treated and polished)

Ha: µ(Unpolished) ≠ µ(Treated and polished)

Where µ(Unpolished) be the population mean of Unpolished stones and µ(Treated and polished) be the population mean of Treated and polished

Level of Significance: α = 0.05

```python
# find the sample means and sample standard deviations for the two samples
print('The mean of Unpolished stones is ' + str(df['Unpolished '].mean()))
print('The mean of Treated and Polished stones is ' + str(df['Treated and Polished'].mean()))
print('The standard deviation of Unpolished stones is ' + str(round(df['Unpolished '].std(),2)))
print('The standard deviation of Treated and Polished stones is ' + str(round(df['Treated and Polished'].std(),2)))
```

Unequal population standard deviations - As the sample standard deviations are different, the population standard deviations may be assumed to be different.

Now, we will calculate the value of T-statistics and p-value, and compare them to reject OR fail to reject the Null hypothesis.

The not equal to sign in the alternative hypothesis indicates that the test is two-tailed and here we will use two independent Sample test.

```python
test_stat, p_value = stats.ttest_ind(df['Unpolished '], df['Treated and Polished'], equal_var = False, alternative = 'two-sided')
print('The p-value is ', p_value)
print('The test_stat is', test_stat)
```

**Conclusion:**

The p-value is less than  level of significance. so we can reject the null hypothesis with 95% confidence, we can say that there is insufficient evidence to say that the mean hardness of the polished and unpolished stones the same.

For testing of normality, **Shapiro-Wilk’s test** is applied to the response variable.

**Shapiro-Wilk’s test**

**Shapiro-Wilk’s test**

We will test the null hypothesis

    H0:  Response follows a normal distribution
against the alternative hypothesis

    Ha:  Response does not follow a normal distribution

The shapiro() function of Scipy will be used to compute the test statistic and p-value.

**Levene’s test**

**Levene’s test**

We will test the null hypothesis

    H0 : All the population variances are equal

against the alternative hypothesis

    Ha : At least one variance is different from the rest

The levene() function of Scipy will be used to compute the test statistic and p-value.

```python
#Assumption 2: Homogeneity of Variance
#import the required function
from scipy.stats import levene
statistic, p_value = levene(df_a1['Response'][df_a1['Dentist']== 1],
                            df_a1['Response'][df_a1['Dentist']== 2],
                            df_a1['Response'][df_a1['Dentist']== 3],
                            df_a1['Response'][df_a1['Dentist']== 4],
                            df_a1['Response'][df_a1['Dentist']== 5])
# find the p-value
print('The p-value is', p_value)
```

As the p-value 0.2565537418543795 is greater than the level of significance, we fail to reject the null hypothesis.

**Let's test whether the assumptions are satisfied or not**
* The populations are normally distributed - No, the normality assumption is verified using the Shapiro-Wilk’s test.
* Population variances are equal - Yes, the homogeneity of variance assumption is verified using the Levene's test.

**Let's find the p-value**

The f_oneway() function of Scipy will be used to compute the test statistic and p-value.



```python
#import the required function
from scipy.stats import f_oneway

# perform one-way anova test
test_stat, p_value = f_oneway(df_a1.loc[df_a1['Dentist'] == 1, 'Response'],
                              df_a1.loc[df_a1['Dentist'] == 2, 'Response'],
                              df_a1.loc[df_a1['Dentist'] == 3, 'Response'],
                              df_a1.loc[df_a1['Dentist'] == 4, 'Response'],
                              df_a1.loc[df_a1['Dentist'] == 5, 'Response'])
print('The test_stat is ' + str(test_stat))
print('The p-value is ' + str(p_value))
```
